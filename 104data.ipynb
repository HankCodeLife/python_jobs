```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import re
```

```python
df = pd.DataFrame()
```

```python
kw = 'python'
baseUrl = 'https://www.104.com.tw/jobs/search/?keyword=' + kw
baseUrl
```

```python
html = requests.get(baseUrl).text
soup = BeautifulSoup(html, 'lxml')
page = int(re.findall(r"\"totalPage\":(\d+)", html)[0])
page
```

```python
# if page >= 15:
#     page = 2
# page
```

```python
for i in range(page):
    url = baseUrl + '&page=' + str(i+1)
    print(url + ' 下載中...')
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'lxml')
    job = soup.select('#js-job-content .job-list-item')
    
    for j in job:
        
        jobName = ''
        url = ''
        company = ''
        companySort = ''
        area = ''
        experiment = ''
        school = ''
        salary = ''
        content = ''
        
        job_link = j.select('.js-job-link')
        if len(job_link) > 0:
            jobName = j.select('.js-job-link')[0].text
            url = j.select('.js-job-link')[0].get('href')
        
        
        list_lnline_li = j.select('ul.b-list-inline')[0].select('li')
        
        if 1 < len(list_lnline_li):
            company = list_lnline_li[1].find('a').text.replace(' ', '')
        
        if 2 < len(list_lnline_li):
            companySort = list_lnline_li[2].text
        
        intro_b_content_li = j.select('ul.job-list-intro.b-content.b-list-inline')[0].select('li')
        
        if 0 < len(list_lnline_li):
            area =intro_b_content_li[0].text
            
        if 1 < len(list_lnline_li):
            experiment =intro_b_content_li[1].text
            
        if 1 < len(list_lnline_li):
            school =intro_b_content_li[2].text
        
        
        if len(j.select('.job-list-tag span')) > 0:
            salary = j.select('.job-list-tag span')[0].text
        
        if len(j.select('.job-list-item__info')) > 0:
            content = j.select('.job-list-item__info')[0].text
        
        data = [{'職位名稱':jobName,
             '工作網址':url,
             '公司名稱':company,
             '公司類別':companySort,
             '工作地點':area,
             '薪資':salary,
             '工作經驗':experiment,
             '學歷':school,
             '工作內容':content}]

        df = df.append(data, ignore_index=True)
        
    
    
#     if i==2: break
df
```

```python
df_copy = df.copy()
```

```python
df_copy.drop_duplicates(subset='工作網址', keep='first', inplace=True)
df_copy
```

```python
df_copy.to_excel('104data.xlsx', index=False)
```
